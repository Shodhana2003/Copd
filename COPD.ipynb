```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
copd=pd.read_csv("/content/drive/MyDrive/COPD.csv")
copd.head()
```

```python
X=copd.drop('label',axis=1)
y=copd['label']
```

```python
columns_to_delete = ['uid', 'class']
X = copd.drop(columns=columns_to_delete)
```

```python
X
```

```python
from sklearn import preprocessing
min_max_scaler=preprocessing.MinMaxScaler()
X_scale=min_max_scaler.fit_transform(X)
print(X_scale)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scale,y,test_size=0.20,random_state=0)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import LeakyReLU
from keras.layers import Dense
```

WITHOUT DROPOUTS

```python
model=Sequential()
model.add(Dense(32,activation='relu',input_shape=(15,)))
model.add(Dense(32,activation='selu'))
model.add(Dense(69,activation='relu'))
model.add(Dense(34,activation='relu'))
model.add(Dense(57,activation='relu'))
model.add(Dense(46,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=18,epochs=2)
```
